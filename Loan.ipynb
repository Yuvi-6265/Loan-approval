# task 2 - data visualization

#Histograms for All Numeric Columns

```python
import matplotlib.pyplot as plt

numeric_cols = df.select_dtypes(include=['int64', 'float64'])

numeric_cols.hist(figsize=(12, 8),
    edgecolor='black')

plt.suptitle("Distribution of Numeric Variables", fontsize=16)
plt.show()
```

#Box Plots: Identify potential outliers and
visualize the spread of data.


```python
numeric_cols = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(12, 6))
sns.boxplot(data=numeric_cols)
plt.title("Box Plots of Numeric Variables")
plt.xticks(rotation=45)
plt.show()
```

#Analyze categorical variables by creating the following plots: Bar Charts: Visualize the frequency distribution of categorical variables.

# Bar chart

```python
categorical_cols = [
    'Gender',
    'Married',
    'Dependents',
    'Education',
    'Self_Employed',
    'Property_Area'
]
```

```python
for col in categorical_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, data=df)
    plt.title(f'Frequency Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

#Pie charts show the proportion of each category.

```python
for col in categorical_cols:
    plt.figure(figsize=(6,6))

    df[col].value_counts().plot(
        kind='pie',
        autopct='%1.1f%%'
    )
```

# Create scatter plots to explore relationships between pairs of numeric variables.

# perform a correlation analysis to identify relationships between numeric variables

```python
#Calculate Correlation Matrix
numeric_cols = df.select_dtypes(include=['int64', 'float64'])

corr_matrix = numeric_cols.corr()

corr_matrix
```

```python
#Visualize Using a Heatmap

plt.figure(figsize=(8,6))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title('Correlation Heatmap')
plt.show()
```

```python
#Stacked Bar Chart

cross_tab = pd.crosstab(
    df['Education'],
    df['Property_Area']
)

cross_tab.plot(
    kind='bar',
    stacked=True,
    figsize=(8,5)
)

plt.title('Education Distribution Across Property Areas')
plt.xlabel('Education')
plt.ylabel('Count')
plt.legend(title='Property Area')
plt.show()
```

# If the dataset contains geographical information, visualize data on a map to identify regional trends

```python
#there is no true geographical information given so we analyse regional trends in it using property_area

plt.figure(figsize=(8,5))

sns.boxplot(
    x='Property_Area',
    y='LoanAmount',
    data=df
)

plt.title('Loan Amount by Property Area')
plt.show()
```

#scatter Plot Colored by Property Area

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x='ApplicantIncome',
    y='LoanAmount',
    hue='Property_Area',
    data=df
)

plt.title('Income vs Loan Amount by Property Area')
plt.show()
```
